## Ejercicio 1 

Dos dados se lanzan una vez y se observa el total obtenido. Usa una simulación para encontrar la probabilidad estimada de que la puntuación sea mayor a 7 o un número par. Una simulación es una repetición del mismo experimento multiples veces para observar su comportamiento:

- Ejecuta el experimento 1000 veces (lanza 2 dados 1000 veces, y suma el número de ambos dados).
- Lleva ua cuenta de los números y las veces que la suma fue mayor a 7 o un número par.
- Divide el número del paso 2 entre el número de iteraciones(1000).

```python
import numpy as np
import random

def dados(n_simulation=1000):
    count=0
    for i in range(n_simulation):

        dado1 = random.randint(1, 6)
        dado2 = random.randint(1, 6)
    
        score = dado1 + dado2
        
       
        if (score % 2 == 0) or (score > 7):
            count += 1

    return count / n_simulation

print(f"La probabilidad de que el número sea mayor 7 o un numero par es: {np.round(dados() * 100, 2)}%")
    
```

## Ejercicio 2

Una caja contiene 10 bolas blancas, 20 bolas rojas y 30 bolas verdes. Si tomamos 5 bolas de la caja reemplazándolas (tomando una bola, anotando el color y luego regresandola a la caja). Queremos saber la probabilidad de:

1. Tomar 3 blancas y 2 rojas.
2. Toas todas del mismo color.

Ejecuta el experimento 1000 veces y calcula las probabilidades que mencionadas.

```python
import numpy as np

caja_bolas = {}

for i in range(60):
    if i < 10:
        caja_bolas[i] = "Blanca"
    elif (i > 9) and (i < 30):
        caja_bolas[i] = "Roja"
    else:
        caja_bolas[i] = "Verde"

print(caja_bolas)
            
def Seleccion_bola(n_simulations=1000):
    count_1 = 0
    count_2 = 0

    for _ in range(n_simulations):
        colors = []

        for _ in range(5):
            colors.append(caja_bolas[np.random.randint(0, 59)])

        colors = np.array(colors)
        
        bolas_blancas = np.sum(colors == "Blanco")
        bolas_rojas = np.sum(colors == "Rojo")
        bolas_verdes = np.sum(colors == "Verde")

        if (bolas_blancas == 3) and (bolas_rojas == 2):
            count_1 += 1
        
        if (bolas_rojas == 5) or (bolas_rojas == 5) or (bolas_verdes == 5):
            count_2 += 1
    
    return count_1 / n_simulations, count_2 / n_simulations

probabilities = Seleccion_bola(1000)

print(f"La probabilidad de tomar una bola blanca y roja es: {np.round(probabilities[0] * 100, 2)}%")
print(f"La probabilidad de unas del mismo color es: {np.round(probabilities[1] * 100, 2)}%")
```
