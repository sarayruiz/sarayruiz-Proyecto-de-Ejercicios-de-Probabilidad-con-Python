## Exercise 1 

Two dices are thrown once and the total score is observed. Use a simulation to find the estimated probability that the total score is even or greater than 7.  A simulation is a repetition of the same experiment multiple times to observe its behavior:

- Run the experiment 1000 times (roll 2 dice 1000 times, and sum the number of both dices).
- Keep track of the number of times that the sum was either greater than 7 or an even number.
- Divide the number from step 2 by the number of iterations (1000).

```python
# TODO

import numpy as np
import random

def dice_rolls(reps):
    rolls_record = 0
    for num in range(reps):
        dice_1 = random.randint(1, 6)
        dice_2 = random.randint(1, 6)

        sum_of_scores = dice_1 + dice_2

        if (sum_of_scores > 7) or (sum_of_scores % 2 == 0):
            rolls_record += 1

    return rolls_record / reps

print(f'{np.round(dice_rolls(1000) * 100, 2)}% Probabilities of scoring even or greater than 7.')
```

## Exercise 2

A box contains 10 white balls, 20 red balls and 30 green balls. If we take 5 balls from the box with replacement (we take the ball, observe what color it is and put it back into the box). We want to know the probability of:

1. Take 3 white and 2 red.
2. Take all of the same color.

Run the experiment 1000 times and calculate the above probabilities.

```python
ball_box = {}

# Create the box of balls
for i in range(60):
    if i < 10:
        ball_box[i] = "White"
    elif (i > 9) and (i < 30):
        ball_box[i] = "Red"
    else:
        ball_box[i] = "Green"

print(ball_box)
            
# TODO

def pulling_balls_out(reps):

    three_white_two_red = 0
    all_same_color = 0

    for num in range(reps):
        record_of_colors = []

        record_of_colors = [ball_box[np.random.randint(0, 60)] for _ in range(5)]

        white = record_of_colors.count("White")
        red = record_of_colors.count("Red")
        green = record_of_colors.count("Green")

        if (white == 3) and (red == 2):
            three_white_two_red += 1
        elif (white == 5):
            all_same_color += 1
        elif (red == 5):
            all_same_color += 1
        elif (green == 5):
            all_same_color += 1

    return three_white_two_red / reps, all_same_color / reps
chance = pulling_balls_out(1000)
print(f"The chances of pulling 3 white and 2 red balls from the box is: {np.round(chance[0] * 100, 2)}%")
print(f"The chances of pulling all same color balls from the box is: {np.round(chance[1] * 100, 2)}%")
```
