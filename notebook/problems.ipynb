### Exercise 1 

Two unbiased dice are thrown once and the total score is observed. Use a simulation to find the estimated probability that the total score is even or greater than 7

We could use a mathematical equation to solve this or we can run the experiment many times, and then count the number of times where the score is even or greater than 7

-Run the experiment 1000 times (roll 2 dice 1000 times, and sum the result)

-Keep track of the number of times that the sum was either greater than 7 or an even number

-Divide the number from step 2 by the number of iterations (1000)


```python
#import libraries
import random


#Funtion to roll the dice.
def lanzar_dados():
    dado1 = random.randint(1, 6)
    dado2 = random.randint(1, 6)
    return dado1 + dado2

tiradas = 1000
exitos = 0

for _ in range(tiradas):
    suma = lanzar_dados()
    if suma > 7 or suma % 2 == 0:
        exitos += 1

probabilidad_estimada = exitos / tiradas
multiplicacion = probabilidad_estimada * 100

print("La probabilidad estimada de que la puntuación total sea par o mayor que 7 es:", multiplicacion, "%")

```

### Exercise 2

A box contains 10 white balls, 20 reds and 30 greens. Draw 5 balls with replacement. What is the probability that:

a. 3 white or 2 red

b. All 5 are the same color


We will use the same system here. We will pick our balls during each round and count the number of times that:

a. 3 white or 2 red 

b. All 5 are the same color.

```python
#importar la biblioteca numpy
import numpy as np

#Diccionario
d = {}

for i in range(60):
    if i < 10:
        d[i] = 'white'
    elif i >= 10 and i < 30:
        d[i] = 'red'
    else:
        d[i] = 'green'

#Initialize important variables
cogidas = 1000
a = 0
b = 0

#Realizar las simulaciones
for i in range(cogidas):
    #Make a list of the colors that we choose
    numeros = np.random.randint(60, size=5)
    
    #Mapear los números a los colores correspondientes en el diccionario
    colores = [d[numero_aleatorio] for numero_aleatorio in numeros]
    
    #Convert it to a numpy array
    colores = np.array(colores)

    #find the number of each taht we picked
    num_blancas = np.count_nonzero(colores == 'white')
    num_rojas = np.count_nonzero(colores == 'red')
    num_verdes = np.count_nonzero(colores == 'green')

    #Keep track if the combination met the criteria
    if num_blancas == 3 and num_rojas == 2:
        a += 1
    if num_blancas == 5 or num_rojas == 5 or num_verdes == 5:
        b += 1
    
#Calcular la probabilidad estimada
prob_a = (a / cogidas) * 100
prob_b = (b / cogidas) * 100

# Imprimir los resultados
print("La probabilidad de obtener 3 bolas blancas o 2 bolas rojas es:", prob_a, "%")
print("La probabilidad de obtener las cinco bolas del mismo color:", prob_b, "%")
```
